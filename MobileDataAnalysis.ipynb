```python
top_company = df['Company Name'].value_counts().head(20)
```

```python
top_company
```

```python
top_company.plot(kind='bar', color='red', figsize=(10,6))
plt.title('Top Mobiles Companies Having Maximum Models')
plt.xlabel('Mobile Companies')
plt.ylabel('Number of Models')
plt.show()
```

**Mobile Companies Analysis:**

1. The mobile company named **OPPO** has more number of models as compared to other mobile companies. 
2. Apple and Huawei sell their mobiles at very high cost. The number of models of Huawei is less as compared to Apple.
3. OnePlus and OPPO mobiles are moderatly expensive mobile phones.
4. Nokia and Infinix mobiles are budget friendly mobile phones.

```python
df['Cost_In_India'] = df['Launched Price (India)']
df['Cost_In_India'] = df['Cost_In_India'].str.replace('INR','')
df['Cost_In_India'] = df['Cost_In_India'].str.replace(',','')
df['Cost_In_India'] = df['Cost_In_India'].str.strip()
df['Cost_In_India'] = df['Cost_In_India'].astype(float)
df['Cost_In_India'].info()
```

```python
df.groupby('Launched Year')['Cost_In_India'].mean().plot(kind='line', color='red', marker='o')
plt.title('Average Price Variation Each Year In India ')
plt.xlabel('Launched Year')
plt.ylabel('Price (INR)')
plt.grid(True)
plt.show()
```

```python
df['Cost_In_Pakistan'] = df['Launched Price (Pakistan)']
df['Cost_In_Pakistan'] = df['Cost_In_Pakistan'].replace(['N/A','Not available'],'')
df['Cost_In_Pakistan'] = df['Cost_In_Pakistan'].str.replace('PKR','')
df['Cost_In_Pakistan'] = df['Cost_In_Pakistan'].str.replace(',','')
df['Cost_In_Pakistan'] = df['Cost_In_Pakistan'].str.strip()
df['Cost_In_Pakistan'] = pd.to_numeric(df['Cost_In_Pakistan'])
df['Cost_In_Pakistan'].info()
```

```python
df.groupby('Launched Year')['Cost_In_Pakistan'].mean().plot(kind='line', color='cyan', marker='o')
plt.grid(True)
plt.title('Average Price Variation Per Year In Pakistan')
plt.xlabel('Launched Year')
plt.ylabel('Price (PKR)')
plt.show()
```

```python
df['Cost_In_USA'] = df['Launched Price (USA)']
df['Cost_In_USA'] = df['Cost_In_USA'].replace(['N/A','Not available'],'')
df['Cost_In_USA'] = df['Cost_In_USA'].str.replace('USD','')
df['Cost_In_USA'] = df['Cost_In_USA'].str.replace(',','')
df['Cost_In_USA'] = df['Cost_In_USA'].str.strip()
df['Cost_In_USA'] = pd.to_numeric(df['Cost_In_USA'])
df['Cost_In_USA'].info()
```

```python
df.groupby('Launched Year')['Cost_In_USA'].mean().plot(kind='line', color='black', marker='o')
plt.grid(True)
plt.title('Average Price Variation Per Year In USA')
plt.xlabel('Launched Year')
plt.ylabel('Price (USD)')
plt.show()
```

```python
df['Cost_In_China'] = df['Launched Price (China)']
df['Cost_In_China'] = df['Cost_In_China'].replace('Not available','')
df['Cost_In_China'] = df['Cost_In_China'].str.replace('CNY','')
df['Cost_In_China'] = df['Cost_In_China'].str.replace(',','')
df['Cost_In_China'] = df['Cost_In_China'].str.strip()
df['Cost_In_China'] = pd.to_numeric(df['Cost_In_China'], errors='coerce')
df['Cost_In_China'].info()
```

```python
df.groupby('Launched Year')['Cost_In_China'].mean().plot(kind='line', marker='o')
plt.title('Average Price Variation Per Year In China')
plt.xlabel('Launched Year')
plt.ylabel('Price (CNY)')
plt.grid(True)
plt.show()
```

```python
df['Cost_In_Dubai'] = df['Launched Price (Dubai)']
df['Cost_In_Dubai'] = df['Cost_In_Dubai'].replace('Not available','')
df['Cost_In_Dubai'] = df['Cost_In_Dubai'].str.replace('AED','')
df['Cost_In_Dubai'] = df['Cost_In_Dubai'].str.replace(',','')
df['Cost_In_Dubai'] = df['Cost_In_Dubai'].str.strip()
df['Cost_In_Dubai'] = pd.to_numeric(df['Cost_In_Dubai'], errors='coerce')
df['Cost_In_Dubai'].info()
```

```python
import pandas as pd
df = pd.read_csv('preprocessedess1.csv', encoding='ISO-8859-1')
#df = df.drop('Processor', axis=1)
df
```

```python
ylist = ['IND_Price','PAK_Price','DUB_Price','CHN_Price','USA_Price']
#xlist = ['RAM','Front Camera','Back Camera','Battery Capacity']
```

```python
X = df.drop(columns=ylist)
#X = df[xlist]
y = df[ylist]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import GradientBoostingRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
rfr = MultiOutputRegressor(RandomForestRegressor(random_state=42, n_estimators=200))
rfr.fit(X_train,y_train)
```

```python
y_train_pred = rfr.predict(X_train)
y_test_pred = rfr.predict(X_test)
print('Training R2:',r2_score(y_train , y_train_pred)*100)
print('Testing R2:',r2_score(y_test , y_test_pred)*100)
```

```python
print('Training MAE:',mean_absolute_error(y_train, y_train_pred)*100)
print('Testing MAE:',mean_absolute_error(y_test, y_test_pred)*100)
```

```python
import numpy as np
print('Training RMSE:',np.sqrt(mean_squared_error(y_train, y_train_pred))*100)
print('Testing RMSE:',np.sqrt(mean_squared_error(y_test, y_test_pred))*100)
```

```python
gbr = MultiOutputRegressor(GradientBoostingRegressor(random_state=42, n_estimators=200))
gbr.fit(X_train,y_train)
```

```python
y_train_pred = gbr.predict(X_train)
y_test_pred = gbr.predict(X_test)
print('Training R2:',r2_score(y_train , y_train_pred)*100)
print('Testing R2:',r2_score(y_test , y_test_pred)*100)
```

```python
print('Training MAE:',mean_absolute_error(y_train, y_train_pred)*100)
print('Testing MAE:',mean_absolute_error(y_test, y_test_pred)*100)
```

```python
import numpy as np
print('Training RMSE:',np.sqrt(mean_squared_error(y_train, y_train_pred))*100)
print('Testing RMSE:',np.sqrt(mean_squared_error(y_test, y_test_pred))*100)
```

```python
lgb = MultiOutputRegressor(LGBMRegressor(random_state=42, n_estimators=200))
lgb.fit(X_train,y_train)
```

```python
y_train_pred = lgb.predict(X_train)
y_test_pred = lgb.predict(X_test)
print('Training R2:',r2_score(y_train , y_train_pred)*100)
print('Testing R2:',r2_score(y_test , y_test_pred)*100)
```

```python
print('Training MAE:',mean_absolute_error(y_train, y_train_pred)*100)
print('Testing MAE:',mean_absolute_error(y_test, y_test_pred)*100)
```

```python
import numpy as np
print('Training RMSE:',np.sqrt(mean_squared_error(y_train, y_train_pred))*100)
print('Testing RMSE:',np.sqrt(mean_squared_error(y_test, y_test_pred))*100)
```
